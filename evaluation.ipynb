```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
usad_file = 'usad_SCADA_Station1_14999_50.dat'
pasad_file = 'pasad_SCADA_Station1_15000_50.dat'

usad_scores = []
true_label = []
pasad_scores = []
pasad_label = []

with open(f'scores/{usad_file}', 'r') as file:
    for line in file:
        values = line.strip().split(";")
        usad_scores.append(float(values[0]))
        true_label.append(float(values[1]))

with open(f'scores/{pasad_file}', 'r') as file:
    for line in file:
        values = line.strip().split(";")
        pasad_scores.append(float(values[0]))
        pasad_label.append(float(values[1]))
```

```python
len(true_label)
```

```python
nums = np.arange(1,len(pasad_scores)+1)
plt.subplot(3,1,1)
plt.plot(nums,pasad_scores)
plt.title(f"PASAD")
plt.ylabel("Departure Score")

for i in range(len(nums) - 1):
    if pasad_label[i] == 1:
        plt.axvspan(nums[i], nums[i+1], color='lightcoral', alpha=0.5)

plt.subplot(3,1,2)
plt.plot(nums, usad_scores)
plt.title("USAD")
plt.ylabel("Preducted Label")

plt.subplot(3,1,3)
plt.plot(nums, true_label)
plt.title("True Value")

plt.tight_layout(pad=.5)

plt.show()
```
